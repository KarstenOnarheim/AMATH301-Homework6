```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

### Part a - Create a meshgrid

You don't need to go through all of these individual parts if you want, it's just to help you organize your work.

```python
##################### Coding Problem 3 ##########################
g = 9.8
L = 11
sigma = 0.12

thetaPrime = lambda v: v
vPrime = lambda theta, v: -g/L * np.sin(theta)  -  sigma * v
odefun = lambda t, p: np.array([thetaPrime(p[1]), vPrime(p[0], p[1])])

v = np.linspace(-3, 3, 25)
theta = v*np.pi

# Create the mesh
theta, v = np.meshgrid(theta, v)

```

### Part b - Create a quiver plot

```python
fig, ax = plt.subplots()
ax.quiver(theta, v, thetaPrime(v), vPrime(theta, v))
plt.xlabel(r'$\theta$')
plt.ylabel("v")
plt.title("dv/d" + r'$\theta$' + " at each point")
plt.xlim(-3*np.pi, 3*np.pi)
plt.ylim(-3, 3)
```

### Part d - Include trajectories.

```python
tvals = np.arange(0, 50 + .01, .01)
sol1 = scipy.integrate.solve_ivp(odefun, [0, 50], [np.pi, .1], t_eval = tvals)
sol2 = scipy.integrate.solve_ivp(odefun, [0, 50], [np.pi, -.1], t_eval = tvals)
sol3 = scipy.integrate.solve_ivp(odefun, [0, 50], [2*np.pi, -3], t_eval = tvals)
sol4 = scipy.integrate.solve_ivp(odefun, [0, 50], [-2*np.pi, 3], t_eval = tvals)

ax.plot(sol1.y[0], sol1.y[1])
ax.plot(sol2.y[0], sol2.y[1])
ax.plot(sol3.y[0], sol3.y[1])
ax.plot(sol4.y[0], sol4.y[1])

fig
```

### Part f - Discussion

#### Part (i) - long-term behavior
The different intial conditions are all leading to the same end condition where theta approaches an integer times 2*pi. In other words, they all approach the same physical position, having rotated around the hinge a different number of times and in different directions. In other words, the pendulum wraps around the top of the hinge if its initial velocity is large enough, and then it eventually settles at the bottom position as a result of gravity.

#### Part (ii) - Comparing two solutions with $\theta_0 = \pi$. 
The two solutions are starting directly above the hinge, then being pushed in slightly different directions, therefore having wrapped around the rod or not, depending on what way it was pushed. This is why they have the different result of 2*pi and 0. The one that started with a positive velocity went to 2*pi because it wrapped, and the other one never wrapped.

#### Part (iii) - Comparing two solutions with equal and opposite $\theta_0$ and $v_0$. 
The one that starts at 2*pi, -3 rotates clockwise 2 full rotations before settling. Having started at 2*pi, it now settles over time to -2*pi because it rotated -4*pi radians. The one that starts at -2*pi, 3 does the exact opposite. It rotates counter-clockwise 2 full times. It starts at -2*pi and rotates 4*pi radians, ending up at 2*pi radians.
